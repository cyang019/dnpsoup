```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import re
```

```python
os.getcwd()
```

```python
logpath = '../../../eulers.log'
os.path.exists(logpath)
```

```python
euler1s = []
euler2s = []
euler3s = []
magic_angles_a, magic_angles_b, magic_angles_g = [], [], []
number_pattern = '-?\d*\.?\d*'
euler_pattern = f'({number_pattern}), ({number_pattern}), ({number_pattern})'
def extract_euler(line):
    found = re.search(euler_pattern, line)
    a = float(found.group(1))
    b = float(found.group(2))
    g = float(found.group(3))
    return a,b,g

with open(logpath, mode='r', encoding='utf-8') as f:
    i = 0
    for line in f:
        if line.startswith('euler'):
            a, b, g = extract_euler(line)
            if line.startswith('euler1:'):
                euler1s.append((a,b,g))
            elif line.startswith('euler2:'):
                euler2s.append((a,b,g))
            else:
                euler3s.append((a,b,g))
            if i % 500 == 0:
                print('.', end='')
            i += 1
        elif line.startswith('[euler]'):
            a, b, g = extract_euler(line)
            magic_angles_a.append(a)
            magic_angles_b.append(b)
            magic_angles_g.append(g)
            if i % 500 == 0:
                print('.', end='')
            i += 1
```

```python
print(len(euler1s))
print(len(euler2s))
print(len(euler3s))
print(len(magic_angles_a))
```

```python
df = pd.DataFrame({
    'a': magic_angles_a,
    'b': magic_angles_b,
    'g': magic_angles_g
})
```
