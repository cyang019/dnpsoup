```python
datadir = '../../../results/v2/NOVEL'
```

```python
filenames = [
    'eH_NOVEL_xband_loop6k_fp.result',
    'eH_NOVEL_xband_loop7k_fp.result',
    'eH_NOVEL_xband_loop100k_fp.result'
]
n = len(filenames)
colors = plt.cm.rainbow(np.linspace(0.1,0.9,n))
labels = [
    'loop 6k',
    'loop 7k',
    'loop 100k'
]

dfs = []
for filename in filenames:
    filepath = os.path.join(datadir, filename)
    df_temp = pd.read_csv(filepath, skiprows=2, header=None)
    dfs.append(df_temp)
df_temp = None

fig = plt.figure(figsize=(10, 6))
for df, c, label in zip(dfs, colors, labels):
    plt.plot(df[0], df[1], linestyle='-', color=c, markerfacecolor='none', label=label)
plt.xlabel('Magnetic Field (T)')
plt.ylabel('Enhancement')
plt.title('NOVEL Field Profile')
plt.legend()
plt.show()
```
