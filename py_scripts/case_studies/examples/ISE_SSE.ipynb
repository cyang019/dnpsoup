### Investigate Euler Angle

```python
labels = [
    'a0b0g0',
    'a45b0g0',
    'a90b0g0',
    'a120b0g0',
    'a160b0g0'
#     'a0b90g0'
]
```

```python
filenames = [
    'eH_gtensor_trityl_a0b90b0_wband_loop4k_xtal_em_fp.result',
    'eH_gtensor_trityl_a0b90b0_wband_loop4k_xtal_a45b0g0_em_fp.result',
    'eH_gtensor_trityl_a0b90b0_wband_loop4k_xtal_a90b0g0_em_fp.result',
    'eH_gtensor_trityl_a0b90b0_wband_loop4k_xtal_a120b0g0_em_fp.result',
    'eH_gtensor_trityl_a0b90b0_wband_loop4k_xtal_a160b0g0_em_fp.result'
#     'eH_gtensor_trityl_a0b90b0_wband_loop4k_xtal_a0b90g0_em_fp.result'
]
```

```python
n = len(filenames)
colors = plt.cm.rainbow(np.linspace(0.1,0.9,n))

dfs = []
for filename in filenames:
    filepath = os.path.join(data_dir, filename)
    df_temp = pd.read_csv(filepath, skiprows=2, header=None)
    dfs.append(df_temp)
df_temp = None

fig = plt.figure(figsize=(10, 6))
for df, c, label in zip(dfs, colors, labels):
    plt.plot(df[0]/1e9, df[1], linestyle='-', color=c, markerfacecolor='none', label=label)
plt.xlabel('Gyrotron Frequency (GHz)')
plt.ylabel('Enhancement')

plt.title('TOP-DNP Field Profile')
plt.legend()
plt.show()
```
